```python
# Showing the realtionship between different effectivenss to music genre
data = {
    'Music Focus Type': ['Lofi', 'Instrumental', 'Pop', 'Rock', 'Ambient Sounds'],
    'Focus Effectiveness 1': [1, 2, 0, 0, 1],
    'Focus Effectiveness 2': [0, 1, 1, 1, 1],
    'Focus Effectiveness 4': [0, 0, 1, 2, 0],
    'Focus Effectiveness 6': [2, 0, 2, 1, 0]
}


df = pd.DataFrame(data)


plt.figure(figsize=(12, 7)) 
bar_width = 0.5 
bar_colors = ['#5B84B1FF', '#FC766AFF', '#5F4B8BFF', '#42EADDFF']  


bottom = [0] * len(df)
for i, col in enumerate(df.columns[1:]):
    plt.bar(
        df['Music Focus Type'], 
        df[col], 
        label=col.replace("Focus Effectiveness ", "Effectiveness "), 
        bottom=bottom, 
        color=bar_colors[i], 
        edgecolor='black',
        width=bar_width
    )
    bottom = [a + b for a, b in zip(bottom, df[col])] 

for index, row in df.iterrows():
    cumulative_value = 0
    for col in df.columns[1:]:
        value = row[col]
        if value > 0:
            plt.text(
                index, 
                cumulative_value + value / 2, 
                str(value), 
                ha='center', 
                va='center', 
                fontsize=10, 
                color='black'
            )
            cumulative_value += value


plt.title("Focus Effectiveness by Music Genre ", fontsize=16, fontweight="bold")
plt.xlabel("Music Focus Type", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(title="Focus Effectiveness", fontsize=10, title_fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.tight_layout()


plt.show()

```
